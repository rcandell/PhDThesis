```python
import numpy as np
import pylab as pl
import matplotlib.pyplot as py

class Radar(object):

    def __init__(self, fig, titles, labels, rect=None):
        if rect is None:
            rect = [0.05, 0.05, 0.95, 0.95]

        self.n = len(titles)
        #self.angles = np.arange(90, 90+360, 360.0/self.n)
        self.angles = [a if a <=360. else a - 360. for a in np.arange(90, 90+360, 360.0/self.n)]
        self.axes = [fig.add_axes(rect, projection="polar", label="axes%d" % i) 
                         for i in range(self.n)]

        self.ax = self.axes[0]
        self.ax.set_thetagrids(self.angles, labels=titles, fontsize=12, weight="bold")

        for ax in self.axes[1:]:
            ax.patch.set_visible(False)
            ax.grid("off")
            ax.xaxis.set_visible(False)

        for ax, angle, label in zip(self.axes, self.angles, labels):
            ax.set_rgrids(range(1, 8), labels=label, angle=angle, fontsize=12)
            ax.spines["polar"].set_visible(False)
            ax.set_ylim(0, 7) 

    def plot(self, values, *args, **kw):
        angle = np.deg2rad(np.r_[self.angles, self.angles[0]])
        values = np.r_[values, values[0]]
        self.ax.plot(angle, values, *args, **kw)
    def fill(self, values, *args, **kw):
        angle = np.deg2rad(np.r_[self.angles, self.angles[0]])
        values = np.r_[values, values[0]]
        self.ax.fill(angle, values, *args, **kw)


fig = pl.figure(figsize=(200, 200))

#titles = [    "Canada", "Australia",     "New Zealand", "Japan", "China", "USA", "Mexico", "Finland", ]



titles = [
    "End-to-end latency in ms", 
    "Update rate in Hz", 
    "\n Payload size \n in Byte", 
    "Date rate per link in bps ", 
    "Data rate per area in bps", 
    "Nodes per area", 
    "\n \n x in m, \n area size: x*x", 
    "Reliability in x \"nines\" \n \n"
]

#labels = [
#    list("abcde"), list("1234567"), list("uvwxy"), 
#    [" ", " ", "$156", "$158", "$160"],
#    list("jklmn"), list("asdfg"), list("qwert"), [" ", "4.3", "4.4", "4.5", "4.6"]
#]

labels = [
    ["1 K", "100", "50", "10", "5", "1", ""],
    ["0.01", "0.1", "1", "10", "100", "1 K", ""],
    ["1", "10", "50", "100", "256", "1 K", ""],
    ["1 K", "10 K", "100 K", "1 M", "10 M", "100 M", ""],
    ["1 M", "10 M", "100 M", "1 G", "10 G", "100 G", "1 T"],
    ["10", "50", "100", "1 K", "10 K", "100 K", ""],
    ["10", "50", "100", "300", "500", "1 K", "10 K"],
    ["3", "4", "5", "6", "7", "8", "9"]
]

radar = Radar(fig, titles, labels)


radar.plot([6, 6,5,4.2,4.2,4,3,4],  lw=1, color="b", alpha=.8, label="a) Discrete automation - Motion control")
#radar.fill([6, 6,5,4.2,4.2,4,3,4],  alpha=.5)
radar.plot([4,5,6,5,7,6,6,2], lw=1, color="g", alpha=.8, label="b) Discrete automation")
radar.plot([3,4,6,6,5,2.9,4,4], lw=1, color="r", alpha=.8, label="c) Process automation - remote control")
radar.plot([5.75,5.5,2.25,3.2,1.8,1.75,1,6], "--", lw=1.2, color="b", alpha=.8, label="d) Industrial printer")
radar.plot([0.5,1.5,2.1,0.4,0.4,2.1,1.5,1], "--", lw=1.2, color="r", alpha=.8, label="e) TE process control")
radar.plot([6.3,6.3,3,4.2,2.25,1.25,1.125,4], "--", lw=1.2, color="g", alpha=.8, label="f) Machine tool")
radar.plot([6,6,5,4.2,4.2,3,6,4], "--", lw=1.2, color="m", alpha=.8, label="g) Mobile robot - cooperative motion control")

#radar.ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.10), fancybox=True, shadow=False, ncol=2)
radar.ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.10), frameon=False, shadow=False, ncol=2)

fig = py.gcf()
fig.set_size_inches(8, 12, forward=True)
fig.savefig('test2png2.png', dpi=400, bbox_inches="tight", pad_inches=1)

tmp = "Done"
tmp
```
